```python
import pandas as pd
```

```python
df1= pd.read_csv('https://raw.githubusercontent.com/gheniabla/datasets/master/companylist-1.csv')
```

```python
df2= pd.read_csv('https://raw.githubusercontent.com/gheniabla/datasets/master/companylist-2.csv')
```

```python
df3= pd.read_csv('https://raw.githubusercontent.com/gheniabla/datasets/master/companylist-3.csv')
```

```python
bigdata = pd.concat([df1, df2,df3], ignore_index=True).sort_values(['Symbol'])
bigdata
```

```python
bigdata.to_csv('merged.csv',index= False)
```

```python
new_index= ['Symbol', 'Name','LastSale', 'MarketCap', 'IPOyear', 'Sector',  'Industry']
```

```python
bigdata_sorted = bigdata.sort_values(['MarketCap'],ascending=False).head(15).reindex(columns = new_index)

bigdata_sorted
```

```python
bigdata_sorted.reset_index(inplace=True)
bigdata_sorted = bigdata_sorted.rename(columns = {'index':'Sequence Number'})
bigdata_sorted.to_csv('marketcap.csv', index= False)
```

```python

symbol1 = input('please a stock symbol to search: ')
symbol1 = symbol1.upper();
dff = bigdata.loc[bigdata['Symbol']== symbol1]
if dff.empty:
    print("I don't finnd this symbol please enter another symbol ")
else:
    dff = dff[new_index]
    print("Stockes with symbol  is: \n" )
    display(dff)
```
